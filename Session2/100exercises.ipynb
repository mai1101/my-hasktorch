83. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)

```haskell
let n83 = 100    -- 元のデータの数
    k83 = 100  -- 試行回数
x83 <- randnIO' [n83] -- 元のデータ（平均0, 分散1の正規分布など）
index83 <- randintIO' 0 n83 [k83, n83] --0から N-1 までのランダムな整数 の行列
let index83' = toType Int64 index83
let x_samples83 = reshape [k83, n83] $ indexSelect 0 (reshape [k83 * n83] index83') x83 --元のデータからindexに基づいてデータk83*n83個をとってくる
    means83 = meanDim (Dim 1) RemoveDim Float x_samples83 -- [k83] 個の平均値を出す
    
let (sortedMeans83,_) = topK k83 (Dim 0) False True means83
    lowIdx83 = round (fromIntegral k83 * 0.025) --2.5%がk83個のうち何番目にくるか？
    highIdx83 = round (fromIntegral k83 * 0.975) --97.5%がk83個のうち何番目にくるか？
    result83 = indexSelect 0 (asTensor ([lowIdx83, highIdx83] :: [Int])) sortedMeans83
print sortedMeans83
print result83
```
